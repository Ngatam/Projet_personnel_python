```python
scan_paths = sorted(glob(os.path.join(base_dir, '*.tif'))) # assume scans are ordered by time?
print(len(scan_paths), 'scans found')
```

# Quick Segmentation and Labeling

```python
from skimage import color
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.feature import canny
from skimage.draw import circle_perimeter, disk

def intersection(cerc0, cerc1):
    #[cx, cy, rayon]
    return ((cerc1[0] - cerc0[0])**2 + (cerc1[1] - cerc0[1])**2)**0.5 < cerc0[-1] + cerc1[-1]

def detect_circles(in_img):
    edges = canny(in_img, sigma=2.3)

    hough_radii = np.arange(13, 90, 3)

    hough_res = hough_circle(edges, hough_radii)
    accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii, total_num_peaks=100) #skimage list of local max -> list of center + accums ???
                                            #domaine de hough, liste rayons recherché, nombre de cercle à chercher
    print(len(radii), "cercles detectés")
    print("rayon moyen:", np.mean(radii))
    print("rayon max:", np.max(radii))
    print("rayon min:", np.min(radii))
    print("rayon écart type:", np.std(radii))

    circleList = np.array(list(zip(cx, cy, radii)))
    #trie  cercles par ordre croissant de rayon
    circleList = circleList[np.apply_along_axis(lambda row: row[-1], 1, circleList).argsort()][::-1]

    #idée: on groupe les cercle similaire ensemble
    #objectif: pouvoir trouver un cercle unique optimale par pore dans l'échantillon
    #similarité entre cercle: distance entre les cercles:
    #  2 cercles similaire = distance Centre-Centre < (somme rayon)
    
    Selections = [[circleList[0]]]
    for cercle in circleList[1:]:
        outer_loop_should_break = False
        for i, selection in enumerate(Selections): #pour chaque cercle déja sélectioné
            if intersection(cercle, selection[0]): #si similaire
                Selections[i].append(cercle) #ajoute à la liste des cercle similaire
                #rapelle: on commence par les rayon les plus grand, donc rayon le plus grand pour le 1er élément des cercles
                outer_loop_should_break = True #on a trouver un cercle similaire, on passe au cercle suivant
                break
        if outer_loop_should_break == False: #pas de similaire troucé, on a un nouveau cercle
            Selections.append([cercle])
            
    ListCercleFinal = [cercles[0] for cercles in Selections] #on souhaite 1 cercle par famille de cercle simialires
    
    print(len(ListCercleFinal), "cercles unique") #on affiche le nouveau nombres de cercles trouvés
    
    cx, cy, radii = zip(*ListCercleFinal) #on récupère liste de positions et rayons pour nos cercles unique

    radii = np.array(radii) #pour la suite plus simple
    
    ## Calcule porosité:
    Surfacetot = 2*np.pi * (np.shape(in_img)[1]/2 - 35)**2 #estimation du rayon de l'échantillon à la main
    Surfaceporo = 2*np.pi* radii@radii.T
    Porosite = Surfaceporo/Surfacetot
    print("Porosité: ", Porosite * 100, "%")

    ## Affichage
    
    img1 = np.zeros(in_img.shape)
    img1 = color.gray2rgb(img1)
    
    for center_y, center_x, radius, (r, g, b, _) in zip(cy, cx, radii, 
                                          plt.cm.nipy_spectral(np.linspace(0,1, len(radii))) # color map
                                         ):
        circy, circx = circle_perimeter(int(center_y), int(center_x), int(radius))
        circy, circx = disk((center_y, center_x), radius)
        img1[circy, circx] += (255.0, 0.0, 0.0)# (r*255, g*255, b*255)
    
    return img1, Porosite
```

```python
fig, m_axs = plt.subplots(4, 3, figsize = (20, 10))
[c_ax.axis('off') for c_ax in m_axs.flatten()]
Poro = []
for i, (c_path, c_ax) in enumerate(zip(scan_paths, m_axs.flatten())):
    c_img = imread(c_path)
    print(c_path)

    c_img = c_img[c_img.shape[0]//2]
    
    rs_img = np.clip(255*(c_img/c_img.max()), 0, 255).astype(np.uint8)
    
    seg_img, p = detect_circles(rs_img)
    Poro.append(p)
    stack_img = (seg_img/seg_img.max()  + color.gray2rgb(c_img/c_img.max()))
    stack_img = stack_img/stack_img.max()
    
    c_ax.imshow(stack_img)
    c_ax.set_title('{:02d} - {}'.format(i, os.path.basename(c_path)))
    print("")

print("Porosité moyenne:", np.mean(np.array(Poro)) * 100, "%")
```
