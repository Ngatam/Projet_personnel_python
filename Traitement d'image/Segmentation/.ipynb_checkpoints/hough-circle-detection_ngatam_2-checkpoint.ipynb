# Imports

```python
#---- Imports ----#
%matplotlib inline
import numpy as np
from skimage.io import imread
from glob import glob
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
base_dir = os.path.join('input')
```

# Recherche des scans dans le dossier

```python
#---- Recherche des scans dans le dossier ----#
scan_paths = sorted(glob(os.path.join(base_dir, '*.tif'))) # assume scans are ordered by time?
print(len(scan_paths), 'scans found')
```

# Scans du dossier lu

```python
#---- Scan du dossier lu ----#
fig, m_axs = plt.subplots(4, 3, figsize = (20, 12))
[c_ax.axis('off') for c_ax in m_axs.flatten()]
for i, (c_path, c_ax) in enumerate(zip(scan_paths, m_axs.flatten())):
    c_img = imread(c_path)
    c_ax.imshow(c_img[c_img.shape[0]//2], cmap = 'gray')
    c_ax.set_title('{:02d} - {}'.format(i, os.path.basename(c_path)))
```

# Condition de contact

```python
#--- Conditions de contact entre les cercles ----#
def condition_de_contact(cx1, cy1, r1, cx2, cy2, r2):
    distance_centre = (abs(cx1 - cx2)**2 + abs(cy1 - cy2)**2)**1/2 # Calcul de la distance entre les 2 centres des cercles 2
    if distance_centre > (r1 + r2) : # Si la distances entre les 2 centre des cercles 1 et 2 est suprieur  la somme des rayons des cercles 1 et 2, les cercles ne se touchent pas
        return False 
        
    else: # Sinon les cercles se touchent en 1 point ou 2 points
        return True
        
```

# Dtection des cercles

```python
#---- Fonction dtection des cercles ----#
from skimage import color
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.feature import canny
from skimage.draw import circle_perimeter, disk
def detect_circles(in_img):
    edges = canny(in_img, sigma=2.3) # Filtre de Canny pour dtecter les contours de l'image, sigma permet de contrler le lissage avant la dtection des contours
    hough_radii = np.arange(10, 90, 4) # Gnre une liste de rayon possibles pour les cercles entre 10 et 90 pixels avec un pas de 2 pixels
    hough_res = hough_circle(edges, hough_radii) # Applique la transforme de Hough sur l'image, la sortie contient les valeurs d'accumulation pour chaque combinaison de centre et de cercles
    accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii, total_num_peaks=100) # La fonction hough_circle_peaks trouve les total_num_peaks meilleurs cercles dtects et renvoie leurs coordonnes et leurs rayons

    img1 = np.zeros(in_img.shape) # Crer une image noire avec les mmes dimensions que l'image de base
    img1 = color.gray2rgb(img1) # Convertit cette image noire en image RGB

    #---- Statistiques ----#
    rayon_max, rayon_min, rayon_moyen, sigma_rayon = np.max(radii), np.min(radii), np.mean(radii), np.std(radii) # Rayon max, rayon min, rayon moyen et cart-type
    print("Le plus grand rayon est de :", rayon_max, " pixels")
    print("Le plus petit rayon est de :", rayon_min, " pixels")
    print("Le rayon moyen est de :", round(rayon_moyen, 3), " pixels")
    print("L'cart-type des rayons est de :", round(sigma_rayon, 3), " pixels")

    #---- Calcul porosit ----#
    Surface_total = 2*np.pi*(np.shape(in_img)[1]/2)**2 # Surface totale de l'chantillon
    Surface_trou = 0
    for rayon_trou in radii:
        Surface_trou+= 2*np.pi*(rayon_trou)**2
        #- print("\nSurface_trou = ", Surface_trou) # Debugg
    Porosite = Surface_trou/Surface_total
    print("La porosit est de : ", round(Porosite*100, 3), "%")

    #---- Test de contact entre les cercles ----#
    len_cx = len(cx)
    CX_trie, CY_trie, R_trie = [], [], [] # Cre les listes des coordonnes des centre et des rayons que l'on va implmenter lors du trie
    for i in range(0, len_cx, 1):
        for j in range(1, len_cx, 1):
            if condition_de_contact(cx[i], cy[i], radii[i], cx[j], cy[j], radii[j]) == True: # Si il y a contact on va ajouter le cercle avec le plus grand rayon
                if radii[i] < radii[j]:
                    CX_trie.append(cx[j])
                    CY_trie.append(cy[j])
                    R_trie.append(radii[j])
                else:
                    CX_trie.append(cx[i])
                    CY_trie.append(cy[i])
                    R_trie.append(radii[i])
            else: # Si il n'y a pas de contact, on ajoute les 2 cercles
                CX_trie.append(cx[j])
                CY_trie.append(cy[j])
                R_trie.append(radii[j])
                CX_trie.append(cx[i])
                CY_trie.append(cy[i])
                R_trie.append(radii[i])
                
    #- print("\nCX_trie = ", CX_trie) # Debugg
    #- print("\nCXY_trie = ", CY_trie) # Debugg
    #- print("\nR_trie = ", R_trie) # Debugg

    for center_x, center_y, radius, (r, g, b, _) in zip(CX_trie, CY_trie, R_trie, plt.cm.nipy_spectral(np.linspace(0,1, len(radii)))): # Pour chaque cercles dtect, le trace sur l'image noire
        circy, circx = circle_perimeter(center_y, center_x, radius) # Dessines les cercles
        #circy, circx = disk((center_y, center_x), radius) # Remplie les cercles que l'on vient de dessiner
        img1[circy, circx] = (r*255, g*255, b*255) # Ajoute la couleurs au pixels des primtres des cercles
        
    return img1
```

# Affichage aprs trie

```python
#---- Affichage aprs trie ----#
fig, m_axs = plt.subplots(4, 3, figsize = (20, 10)) # Crer une figure avec des sous-figure (sub-plot)
[c_ax.axis('off') for c_ax in m_axs.flatten()]

for i, (c_path, c_ax) in enumerate(zip(scan_paths, m_axs.flatten())): # Boucle pour passer par toutes les images du dossier choisi
    print("\nTraitement de l'image :", i+1)
    c_img = imread(c_path) # Lit l'image actuelle
    c_img = c_img[c_img.shape[0]//2] #
    rs_img = np.clip(255*(c_img/c_img.max()), 0, 255).astype(np.uint8)
    seg_img = detect_circles(rs_img)
    stack_img = np.concatenate([plt.cm.bone(c_img/c_img.max())[:, :, :3], seg_img], 1)
    c_ax.imshow(stack_img, cmap = 'gray')
    c_ax.set_title('{:02d} - {}'.format(i, os.path.basename(c_path)))
```
