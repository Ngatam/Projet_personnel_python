```python
%matplotlib inline
import numpy as np
from skimage.io import imread
from glob import glob
import os
import matplotlib.pyplot as plt
base_dir = os.path.join('input')
```

```python
scan_paths = sorted(glob(os.path.join(base_dir, '*.tif'))) # assume scans are ordered by time?
print(len(scan_paths), 'scans found')
```

# Show previews of the scans


## Middle Axial Slice

```python
fig, m_axs = plt.subplots(4, 3, figsize = (20, 12))
[c_ax.axis('off') for c_ax in m_axs.flatten()]
for i, (c_path, c_ax) in enumerate(zip(scan_paths, m_axs.flatten())):
    c_img = imread(c_path)
    c_ax.imshow(c_img[c_img.shape[0]//2], cmap = 'gray')
    c_ax.set_title('{:02d} - {}'.format(i, os.path.basename(c_path)))
```

# Quick Segmentation and Labeling

```python
from skimage import color
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.feature import canny
from skimage.draw import circle_perimeter, disk
def detect_circles(in_img):
    edges = canny(in_img, sigma=2) # Filtre de Canny pour détecter les contours de l'image, sigma permet de contrôler le lissage avant la détection des contours
    hough_radii = np.arange(10, 90, 4) # Génère une liste de rayon possibles pour les cercles entre 10 et 90 pixels avec un pas de 2 pixels
    hough_res = hough_circle(edges, hough_radii) # Applique la transformée de Hough sur l'image, la sortie contient les valeurs d'accumulation pour chaque combinaison de centre et de cercles
    accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii, total_num_peaks=100) # La fonction hough_circle_peaks trouve les total_num_peaks meilleurs cercles détectés et renvoie leurs coordonnées et leurs rayons

    img1 = np.zeros(in_img.shape) # Créer une image noire avec les mêmes dimensions que l'image de base
    img1 = color.gray2rgb(img1) # Convertit cette image noire en image RGB
    for center_y, center_x, radius, (r, g, b, _) in zip(cy, cx, radii, plt.cm.nipy_spectral(np.linspace(0,1, len(radii)))  color map): # Pour chaque cercles détecté, le trace sur l'image noire
        circy, circx = circle_perimeter(center_y, center_x, radius) # Dessines les cercles
        #circy, circx = disk((center_y, center_x), radius) # Remplie les cercles que l'on vient de dessiner
        img1[circy, circx] = (r*255, g*255, b*255) # Ajoute la couleurs au pixels des périmètres des cercles
    return img1 # Renvoie l'image finale
```

```python
fig, m_axs = plt.subplots(4, 3, figsize = (20, 10))
[c_ax.axis('off') for c_ax in m_axs.flatten()]
for i, (c_path, c_ax) in enumerate(zip(scan_paths, m_axs.flatten())):
    c_img = imread(c_path)
    c_img = c_img[c_img.shape[0]//2]
    rs_img = np.clip(255*(c_img/c_img.max()), 0, 255).astype(np.uint8)
    seg_img = detect_circles(rs_img)
    stack_img = np.concatenate([plt.cm.bone(c_img/c_img.max())[:, :, :3], seg_img], 1)
    c_ax.imshow(stack_img, cmap = 'gray')
    c_ax.set_title('{:02d} - {}'.format(i, os.path.basename(c_path)))
```
